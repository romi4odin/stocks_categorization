```python
import pandas as pd
import os
list_files = []
for file in os.listdir():
  if file.endswith('.csv'):    
  
  # you can change '.csv' to any other type of file, but take into account the pd.read_csv 
  # and change it to a pandas function that helps you read that type of file
  # e.g pd.read_excel in case it is a .xlsx file
  # or pd.read_json in case it is a .json file
    
    df = pd.read_csv(file)
    
    # you can add any piece of code inside this for loop in case
    # you want to create a cleaning pipeline
    
    list_files.append(df)

df = pd.concat(list_files, axis=0, ignore_index=True)
```

How many Unique patterns that exist in the historical stock data set, based on fluctuations in price

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import (
    kelbow_visualizer,
    silhouette_visualizer,
    intercluster_distance,
)
elbow_k = kelbow_visualizer(KMeans(random_state=4), values, k=(2, 10))
kmeans = KMeans(n_clusters=elbow_k.elbow_value_, n_init=10, max_iter=1000)
label_pred = kmeans.fit_predict(values)
print("The best number of cluster is: ", elbow_k.elbow_value_)
```

Identify which all stocks are moving together and which all stocks are different from each other.

```python
# Apply PCA to identify the stocks correlated with each other
df.head()
```

```python
import numpy as np
dt = df.copy()
dt.drop(['DATE','SP500'],inplace=True,axis=1)

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# convert data into numpy arrays
X = dt.values

# Scaling the values
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

pca = PCA(n_components=150)
pca.fit(X_scaled)

# the amount of variance that each PCA explains
var = pca.explained_variance_ratio_

# cumulative varince explains
var1 = np.cumsum(np.round(var, decimals=4)*100)
print(var1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(var1)
```

```python
# Looking at above plot I can consider 25 variables 
pca = PCA(n_components =25)
X1 = pca.fit_transform(X_scaled)
print(X1)
```

```python
print('Number of PCA :',len(pca.components_))
print(abs(pca.components_))
```

```python
comp = pd.DataFrame(pca.components_, columns = dt.columns)
comp.head()
```
